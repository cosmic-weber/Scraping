```python
# Importing important packages
import pandas as pd
import time
import os
from selenium import webdriver
```

```python
#importing packages from selenium

from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
# using the url here to retieve jobs by applying certain filter

url = 'https://www.linkedin.com/jobs/search?keywords=Software%20Engineer&location=Indore%2C%20Madhya%20Pradesh%2C%20India&locationId=&geoId=101389470&f_TPR=&distance=25&f_E=2%2C1&position=1&pageNum=0'
```

```python
driver = webdriver.Chrome(executable_path=r'C:\\Users\\Asus\\Desktop\\chromedriver.exe')
driver.implicitly_wait(10)
driver.get(url)

#ignore the below warnings
```

```python
#here finding the total number of jobs count
job_count = driver.find_elements_by_class_name('results-context-header__job-count')[0].text
```

```python
type(job_count)
```

```python
job_count = int(job_count)
job_count
```

```python
#scrolling through all the jobs and if see more jobs button came then scroll through by clicling on it i.e. infinite scrolling

job = 2

while job <= int((job_count+200))/25+1:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    job += 1
    
    try:
        send = driver.find_element_by_xpath("//button[@aria-label='Load more results']")
        driver.execute_script("arguments[0].click();", send)
        time.sleep(3)
        
    
    except:
        pass
        time.sleep(5)
```

```python
company_name, job_designation = [], []
```

```python
# finding the company's name and appending it to the list

try:
    for i in range(job_count):
        company = driver.find_elements_by_class_name('base-search-card__subtitle')[i].text
        company_name.append(company)
        
except IndexError:
    print('No')
```

```python
len(company_name)
```

```python
try:
    for i in range(job_count):
        title = driver.find_elements_by_class_name('base-search-card__title')[i].text
        job_designation.append(title)
        
except IndexError:
    print('No')
```

```python
len(job_designation)
```

```python
# Now creating dataframe for comapny'name and job title
company = pd.DataFrame(company_name, columns=['company'])
title = pd.DataFrame(job_designation,columns=['job-title'])
```

```python
company
```

```python
title
```

```python
#joining the two dataframe
final = company.join(title)
```

```python
final.to_csv('C:\\Users\\Asus\\Desktop\\linkedin.csv') #saving our data onto csv file
```

```python
# Now finding the jobs links

jobList = driver.find_elements_by_class_name('base-card__full-link')
hrefList = []
for e in jobList:
    hrefList.append(e.get_attribute('href'))
```

```python
job_links = pd.DataFrame(hrefList, columns=['joblinks'])
job_links
```
